```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/MLOM/Labs/Lab 09/diabetes.csv")
data
```

declare X (features) and y (labels) arrays

```python
#data preperation
x = data.iloc[:,:8].values
```

```python
x
```

```python
y = data.iloc[:,8]
```

```python
y
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
# Create an SVC model
model = SVC(C=1.0, kernel = 'linear')

#if linear separatable : linear
#if non linear :  poly
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
# Evaluate the model's accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

## To find Confusion Matrix and Classification Report

```python
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

```python
#for cross validation
scores = cross_val_score(model, X_train, y_train, cv=5)  # 5-fold cross-validation
```

```python
scores
```

```python
#Calculate the average accuracy and any other relevant statistics from the scores
mean_accuracy = scores.mean()
std_deviation = scores.std()
```
